```python
import matplotlib.pyplot as plt
from sklearn import datasets
iris = datasets.load_iris()
data = iris.data
x = [2, 6]
y = [15, 8.5]
plt.plot(x, y)

plt.xlabel("x axis")

plt.ylabel("y axis")

plt.title("Distribution Of the iris datasets ")

plt.grid()
```
